```python
import pandas as pd  
from skimage.measure import regionprops, regionprops_table

'''segmented_intensity = copy.deepcopy(dapi[0])
segmented_intensity[local_thresh_segmented[2] ==0] = 0'''

rprops_table = regionprops_table(init_segmented, intensity_image=subtracted, properties=('label', 'area', 'bbox_area', 'convex_area', 'eccentricity', 'equivalent_diameter', 'euler_number', 'filled_area', 'extent',
                                                                       'major_axis_length', 'minor_axis_length', 'moments_central', 'moments_hu', 'perimeter','orientation',
                                                                       'solidity', 'weighted_moments_central', 'weighted_moments_hu', 'inertia_tensor', 'inertia_tensor_eigvals', 'moments', 'max_intensity', 'mean_intensity', 'min_intensity'))


'''rprops_table = regionprops_table(init_segmented, intensity_image=subtracted, properties=('label', 'area', 'convex_area', 'eccentricity', 'equivalent_diameter', 'filled_area', 'extent',
                                                                       'major_axis_length', 'minor_axis_length', 'moments_central', 'moments_hu', 'perimeter',
                                                                       'solidity', 'weighted_moments_central', 'inertia_tensor', 'moments', 'max_intensity', 'mean_intensity', 'min_intensity'))

rprops_table = regionprops_table(init_segmented, intensity_image=subtracted, properties=('label', 'area', 'convex_area', 'eccentricity', 'euler_number', 'extent', 'filled_area', 'inertia_tensor'
                                                                                         'equivalent_diameter', '''
#'weighted_moments' 'inertia_tensor', 'moments', 'max_intensity', 'mean_intensity', 'min_intensity',


data = pd.DataFrame(rprops_table)  
#data.iloc[:,50:60]

data = data.set_index("label")
data
```

```python
from sklearn.cluster import KMeans


df = data
zscored = (df - df.mean())/df.std(ddof=0)
kmeans = KMeans(n_clusters=6, init='k-means++', random_state = 5).fit(zscored)
#centroids = kmeans.cluster_centers_
print(kmeans.labels_)
```

```python
for i in zscored.index.values[kmeans.labels_ == 1][1:5]:
    tmp = copy.deepcopy(init_segmented)
    tmp[tmp != i] = 0
    fig = plt.figure(figsize = (20,20))
    plt.imshow(tmp)
    plt.figure(i+1)
plt.show()
```

```python
tmp = copy.deepcopy(init_segmented)

cluster_labels = []
test = []
for i in np.unique(kmeans.labels_):
    cluster_labels.append(zscored.index.values[kmeans.labels_ == i])
    test.append(np.isin(tmp, cluster_labels[i]))
    tmp[test[i]] = 11000 - (i + 1) * 1000

#test = np.isin(tmp, cluster_zero_labels)
fig = plt.figure(figsize = (30,30))
plt.imshow(tmp)
plt.show()
```
