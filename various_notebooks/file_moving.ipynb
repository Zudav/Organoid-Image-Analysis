Used to move images from one folder to another. I wrote this to create a data set that contains only those organoids for which all cycles are present. The images that this would filter out were moved to a backup location rather than deleted.  
**Note**: This script is no longer useful because other scripts further up the pipeline have been adjusted such that the filtering step occurs much earlier. This script is only still here for documentation of what I did during my project

```python
from pathlib import Path
```

```python
folder = Path(r"/links/groups/treutlein/DATA/imaging/PW/4i/plate14_aligned_UZH")
destination = Path(r"/links/groups/treutlein/DATA/imaging/PW/4i/plate14_aligned_UZH_backup")

# Cycle names
cycles = ["cycle0", "cycle1", "cycle2", "cycle3", "cycle4", "cycle5",
         "cycle5_0", "cycle6", "cycle7", "cycle8", "cycle9", "cycle10",
         "cycle10_0", "cycle11", "cycle12", "cycle13", "cycle14", "cycle15",
         "cycle15_0", "cycle16", "cycle17", "cycle18", "cycle19", "cycle20",
         "cycle20_0", "cycle21", "cycle1_2", "cycle1_3"]

# Wells to be removed
wells = ["A01", "A02", "A06", "A10", "B10", "B13", "B14", "B15", "B19", "B24", 
         "C05", "C06", "C11", "C12", "C15", "C16", "C17", "C18", "C19", "C20", 
         "C21", "C22", "C23", "D02"]

experiment_name = "plate14"
nchan = 4  # number of channels per image
```

```python
for cycle in cycles:
    print("\n{}".format(cycle))
    for well in wells:
        for chan in range(1, nchan+1):
            filename = "{}_{}_C{:02d}.tif".format(experiment_name, well, chan)
            in_path = Path(folder/cycle/filename)
            out_path = Path(destination/cycle/filename)
            if in_path.is_file():  # Check if input file exists
                if out_path.is_file():  # See if output file already exists
                    #print("Output already exists")
                    continue
                else:
                    # Create folder for cycle if it doesn't exist yet
                    Path(destination/cycle).mkdir(parents=True, exist_ok=True)
                    in_path.rename(str(out_path))
            else:  # If input file doesn't exist
                #print("No file found at path\n{}".format(str(in_path)))
                continue
```
